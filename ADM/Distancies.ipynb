# üìè **Distance Metrics Overview**

Distance metrics are mathematical tools used to measure the similarity or dissimilarity between two points, vectors, or sets. Each metric has its unique application, depending on the problem domain and data structure.

---

### 1Ô∏è‚É£ **Euclidean Distance** (L2 Norm)
- Measures the **straight-line distance** between two points in Euclidean space.
- It is calculated as:

$$
d(x, y) = \sqrt{\sum_{i=1}^{n}(x_i - y_i)^2}
$$

- **Use Case:** Commonly used in clustering (e.g., K-Means), image processing, and regression models.

**Example:** Distance between points (1,2) and (4,6):  
$$
d = \sqrt{(4-1)^2 + (6-2)^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$


```python
import matplotlib.pyplot as plt
import math
import matplotlib.pyplot as plt

# Sample Data Points
point1 = (1, 2)
point2 = (4, 6)
```

```python
def euclidean_distance(point1, point2):
    return math.sqrt(sum((p1 - p2)**2 for p1, p2 in zip(point1, point2))) # Euclidean Distance Formula

plt.figure(figsize=(12, 8))
plt.plot([point1[0], point2[0]], [point1[1], point2[1]], 'o-', label='Euclidean Path', color='blue')
plt.text((point1[0] + point2[0]) / 2, (point1[1] + point2[1]) / 2,
         f'Euclidean: {euclidean_distance(point1, point2):.2f}', color='blue')
```

### 2Ô∏è‚É£ **Manhattan Distance** (L1 Norm)
- Measures the distance by summing the **absolute differences** of their coordinates.
- It is calculated as:

$$
d(x, y) = \sum_{i=1}^{n}|x_i - y_i|
$$

- **Use Case:** Often used in grid-based problems, like routing in city block models.

**Example:** Distance between points (1,2) and (4,6):  
$$
d = |4-1| + |6-2| = 3 + 4 = 7
$$


```python
def manhattan_distance(point1, point2):
    return sum(abs(p1 - p2) for p1, p2 in zip(point1, point2)) # Manhattan Distance Formula
```

```python
plt.plot([point1[0], point1[0], point2[0]], [point1[1], point2[1], point2[1]],
         'o--', label='Manhattan Path', color='orange')
plt.text((point1[0] + point2[0]) / 2, point1[1],
         f'Manhattan: {manhattan_distance(point1, point2):.2f}', color='orange')
```

### 3Ô∏è‚É£ **L‚àû Norm (Chebyshev Distance)**
- Measures the **maximum absolute difference** between corresponding coordinates.
- It is calculated as:

$$
d(x, y) = \max_{i}|x_i - y_i|
$$

- **Use Case:** Used in chessboard problems and real-time strategy games for maximum movement across grids.

**Example:** Distance between points (1,2) and (4,6):  
$$
d = \max(|4-1|, |6-2|) = \max(3, 4) = 4
$$

```python
def chebyshev_distance(point1, point2):
    return max(abs(p1 - p2) for p1, p2 in zip(point1, point2)) # Chebyshev Distance Formula

plt.plot([point1[0], point2[0]], [point1[1], point1[1]], 'o--', label='Chebyshev Path', color='green')
plt.text(point2[0], point1[1], f'Chebyshev: {chebyshev_distance(point1, point2):.2f}', color='green')

```

Alltogether:

```python
# Plot Points and Distances
plt.figure(figsize=(12, 8))
plt.plot([point1[0], point2[0]], [point1[1], point2[1]], 'o-', label='Euclidean Path', color='blue')
plt.text((point1[0] + point2[0]) / 2, (point1[1] + point2[1]) / 2,
         f'Euclidean: {euclidean_distance(point1, point2):.2f}', color='blue')

# Manhattan Path
plt.plot([point1[0], point1[0], point2[0]], [point1[1], point2[1], point2[1]],
         'o--', label='Manhattan Path', color='orange')
plt.text((point1[0] + point2[0]) / 2, point1[1],
         f'Manhattan: {manhattan_distance(point1, point2):.2f}', color='orange')

# Chebyshev Path
plt.plot([point1[0], point2[0]], [point1[1], point1[1]], 'o--', label='Chebyshev Path', color='green')
plt.text(point2[0], point1[1], f'Chebyshev: {chebyshev_distance(point1, point2):.2f}', color='green')

# Highlight Points
plt.scatter(*point1, color='red', label='Point 1')
plt.scatter(*point2, color='purple', label='Point 2')

# Labels and Legend
plt.title('Distance Metrics Visualization')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

### 4Ô∏è‚É£ **Jaccard Distance**
- Measures the **dissimilarity** between two sets.
- It is calculated as:

$$
d(A, B) = 1 - \frac{|A \cap B|}{|A \cup B|}
$$

- **Use Case:** Used in text analysis, document comparison, and clustering with categorical data.

**Example:**  
- Set A = {1, 2, 3}, Set B = {2, 3, 4}  
$$
d = 1 - \frac{|{2, 3}|}{|{1, 2, 3, 4}|} = 1 - \frac{2}{4} = 0.5
$$




```python
set1 = {1, 2, 3}
set2 = {2, 3, 4}
```

```python
# üìù Jaccard Distance
def jaccard_distance(set1, set2):
    if not set1 and not set2: # If both sets are empty
        return 0
    intersection = len(set1 & set2) # Intersection of two sets
    union = len(set1 | set2) # Union of two sets
    return 1 - (intersection / union) # Jaccard Distance Formula

jaccard = jaccard_distance(set1, set2)
print(f'Jaccard Distance: {jaccard:.2f}')
```

### 5Ô∏è‚É£ **Cosine Distance**
- Measures the **angle between two vectors** in multi-dimensional space.
- It is calculated as:

$$
d(x, y) = 1 - \frac{x \cdot y}{||x|| ||y||}
$$

- **Use Case:** Used in text mining, natural language processing, and clustering high-dimensional data.

**Example:**  
- Vectors: $x = (1, 2)$, $y = (2, 3)$  
$$
\text{Cosine Similarity} = \frac{(1)(2) + (2)(3)}{\sqrt{1^2 + 2^2} \sqrt{2^2 + 3^2}}
$$

- Cosine Distance = 1 - Cosine Similarity

```python
vector1 = [1, 2]
vector2 = [2, 3]
```

```python
def cosine_distance(vector1, vector2):
    dot_product = sum(p1 * p2 for p1, p2 in zip(vector1, vector2)) # Dot Product of two vectors
    magnitude1 = sum(p1**2 for p1 in vector1)**0.5 # Magnitude of Vector 1 (first part of the denominator)
    magnitude2 = sum(p2**2 for p2 in vector2)**0.5 # Magnitude of Vector 2 (second part of the denominator)
    if magnitude1 == 0 or magnitude2 == 0: # If any of the magnitudes is zero,
        return 0 # return 0 to avoid division by zero
    return 1 - (dot_product / (magnitude1 * magnitude2)) # Cosine Distance Formula

cosine = cosine_distance(vector1, vector2)
print(f'Cosine: {cosine:.2f}')
```

### 6Ô∏è‚É£ **Edit Distance (Levenshtein Distance)**
- Measures the **minimum number of operations** (insertion, deletion, substitution) to transform one string into another.
- **Use Case:** Used in spell checkers, DNA sequencing, and natural language processing.

**Example:**  
- String 1: `kitten`, String 2: `sitting`  
Operations:  
1. k ‚Üí s (substitution)  
2. e ‚Üí i (substitution)  
3. Add g  
Edit Distance = 3

```python
string1 = "kitten"
string2 = "sitting"
```

```python
def edit_distance(str1, str2):
    m, n = len(str1), len(str2) # Length of both strings
    dp = [[0] * (n + 1) for _ in range(m + 1)] # Create a 2D list with m+1 rows and n+1 columns to store the subproblems
    
    for i in range(m + 1): # Iterate through the rows
        for j in range(n + 1): # Iterate through the columns
            # Dealing with Hard Cases:
            if i == 0: # If the first string is empty
                dp[i][j] = j # Add all characters of the second string
            elif j == 0: # If the second string is empty
                dp[i][j] = i # Add all characters of the first string
            elif str1[i-1] == str2[j-1]: # If the characters are the same
                dp[i][j] = dp[i-1][j-1] # No operation is needed
            else:
                # Main operation
                dp[i][j] = 1 + min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1]) # Find the minimum of three operations: insert, delete, replace
    return dp[m][n] 

edit = edit_distance(string1, string2)
print(f'Edit Distance: {edit}')
```
