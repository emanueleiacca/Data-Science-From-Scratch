With a byte we
can represent $2^8$ = 256 different numbers

With $n$ bits we can represent $2^n$ different numbers. Therefore,
with 1 byte we can represent numbers from 0 to 255 and so on

Binary (Base-2): Uses only 0 and 1. Example: 13 (Decimal) → 1101 (Binary)

Hexadecimal (Base-16): Uses 0-9 and A-F. Example: 13 (Decimal) → D (Hexadecimal)


```python
# Converting Decimal to Binary and Hexadecima
# Decimal number
num = 13

# Convert to binary
binary_rep = bin(num)
print(f"Binary Representation of {num}: {binary_rep}") # 0b prefix means binary

# Convert to hexadecimal
hex_rep = hex(num)
print(f"Hexadecimal Representation of {num}: {hex_rep}") # 0x prefix means hexadecimal
```

Assume that we use 8bits to represent numbers, the 0000 0000 to 0111 1111 are
all positive numbers (and zero) and 1000 0000 to 1111 1111 are all negative (the space in between is only
for readability

```python
# Unsigned and Signed Integer Representation

# Positive number (13 in 8-bit binary)
pos_num = 13
print(f"Positive Integer: {pos_num}")
print(f"Binary Representation: {format(pos_num, '08b')}")  # 8-bit binary

# Negative number (-13 in 8-bit two's complement)
neg_num = -13
# Two's complement for negative number (limit to 8 bits)
binary_neg_num = format(neg_num & 0xFF, '08b')
print(f"Negative Integer: {neg_num}")
print(f"Two's Complement Binary Representation (8-bit): {binary_neg_num}")
```

    Function MaxFind(A):

        Input: A - List of n elements
        Output: Maximum value in A
        
        max ← A[1]
        for i = 2 to n:
            if A[i] > max:
                max ← A[i]
        return max


Big O (O) describes the upper bound of an algorithm's growth rate.

It focuses on how the runtime grows as the input size (n) increases.


```python
def find_max(arr):
    """
    Find the maximum value in a list.
    """
    max_val = arr[0]
    for num in arr[1:]:
        if num > max_val:
            max_val = num
    return max_val

# Example Usage
nums = [3, 5, 1, 9, 2]
print(f"Maximum Value: {find_max(nums)}")
```

- Initialization (Line 4): Runs once → O(1)
- Loop (Lines 5–8): Runs n-1 times → O(n)
- Return Statement (Line 9): Runs once → O(1)

The function loops through the list once (O(n) complexity).


```python
import numpy as np
import matplotlib.pyplot as plt

# Input sizes from 1 to 1000
n = np.arange(1, 1001)

# Define complexity functions
log_n = np.log(n)
linear = n
n_log_n = n * np.log(n)
n_squared = n**2
two_power_n = 2**(n / 50)  # Scale down 2^n for visualization

# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(n, log_n, label='O(log n)')
plt.plot(n, linear, label='O(n)')
plt.plot(n, n_log_n, label='O(n log n)')
plt.plot(n, n_squared, label='O(n^2)')
plt.plot(n, two_power_n, label='O(2^n)')

# Add labels and legend
plt.title('Big O Complexity Visualization')
plt.xlabel('Input Size (n)')
plt.ylabel('Number of Steps')
plt.legend()
plt.yscale('log')  # Use logarithmic scale for better visualization
plt.grid(True)
plt.show()
```

Naive PseudoCode:

    Function PrefixSum(A):

        Input: A - Array of size n
        Output: S - Prefix-sum array of A
        for i = 1 to n:
            S[i] ← 0
            for j = 1 to i:
                S[i] ← S[i] + A[j]
        end for
        return S


- Outer loop runs n times.
- Inner loop runs i times (1, 2, 3, ..., n).


Optimized PseudoCode:

    Function PrefixSumFast(A):

        Input: A - Array of size n
        Output: S - Prefix-sum array of A
        S[1] ← A[1]
        for i = 2 to n:
            S[i] ← S[i − 1] + A[i]
        end for
        return S


- Loop runs n-1 times.
- Each iteration performs a constant-time addition.

### NP-Completeness

- P (Polynomial Time): Problems that can be solved efficiently (in polynomial time) by an algorithm. Example: Finding the maximum value in an array.
- NP (Nondeterministic Polynomial Time): Problems where a given solution can be verified in polynomial time, even if finding the solution might not be efficient (Sudoku verification).

Key Insight: Every problem in P is also in NP, but not every problem in NP is proven to be in P.



### Data Structure

A data structure is a way to store, organize, and manage data efficiently for specific operations (e.g., searching, sorting, inserting, or deleting data).

Different data structures are suited for different operations:

- Linked Lists: Efficient for dynamic insertion/deletion.

```python
from collections import deque

# Create a Queue
queue = deque()

# Enqueue elements
queue.append(3)
queue.append(5)

# Peek front element
print("Front element:", queue[0]) 

# Dequeue elements
print("Removed:", queue.popleft()) 
print("Removed:", queue.popleft())  # Queue is now empty
```

```python
# Create a Stack
stack = []

# Push elements
stack.append(10)
stack.append(20)

# Peek top element
print("Top element:", stack[-1])  

# Pop elements
print("Removed:", stack.pop())  
print("Removed:", stack.pop())  # Stack is now empty
```

### Dynamic Programming

 is an optimization technique used to solve complex problems by breaking them into smaller subproblems.
These subproblems are solved once and their solutions are stored to avoid redundant computations.

```python
def edit_distance_recursive(S, T): 
    i, j = len(S), len(T) 
    
    # Base cases
    if i == 0:
        return j  # Cost = length of T
    if j == 0:
        return i  # Cost = length of S
    
    # If characters match, move diagonally
    if S[-1] == T[-1]:
        return edit_distance_recursive(S[:-1], T[:-1]) # No cost
    else:
        # Otherwise, calculate minimum cost
        return 1 + min(
            edit_distance_recursive(S[:-1], T),  # Remove from S
            edit_distance_recursive(S, T[:-1])   # Remove from T
        )
```

Overlapping subproblems cause repeated calculations

```python
def edit_distance_dp(S, T):
    n, m = len(S), len(T)
    dp = [[0] * (m + 1) for _ in range(n + 1)] # stores the edit distance
    
    # Initialize first row and column
    for i in range(n + 1):
        dp[i][0] = i  # Cost to delete all characters from S
    for j in range(m + 1):
        dp[0][j] = j  # Cost to delete all characters from T
    
    # Fill the DP table
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if S[i - 1] == T[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]  # Characters match
            else:
                dp[i][j] = 1 + min(
                    dp[i - 1][j],    # Remove from S
                    dp[i][j - 1],    # Remove from T
                    dp[i - 1][j - 1] # Substitute
                )
    
    # The final answer is in the bottom-right corner
    return dp[n][m]
```

```python
S = "ALGORITHMS"
T = "LOGARITHM"

print("Recursive Edit Distance:", edit_distance_recursive(S, T))
print("DP Edit Distance:", edit_distance_dp(S, T))
```
