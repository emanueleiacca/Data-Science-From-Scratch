# üìö **Graph Theory**

Graph theory serves as the mathematical foundation for analyzing **networks and relationships** between objects.

üìñ **Definitions in Graph Theory**

‚û°Ô∏è **Vertices (Nodes):** Fundamental units representing objects (e.g., people in a social network).  
‚û°Ô∏è **Edges:** Connections between nodes (e.g., friendships in social networks).  
‚û°Ô∏è **Directed vs. Undirected Graphs:**  
- **Directed:** Edges have a direction (e.g., follower-following relationship), so they indicate one-way relationships ($A \rightarrow B$).  
- **Undirected:** Edges have no direction (e.g., friendship on Facebook) so Relationships are two-way ($A - B$).  
‚û°Ô∏è **Weighted Graphs:** Edges have weights representing the strength or cost of the connection  (e.g., distance, cost).  

‚û°Ô∏è **Loops and Multiple Edges:**  
- **Loop:** An edge connecting a node to itself.  
- **Multiple Edges:** More than one edge between two nodes.

‚û°Ô∏è **Simple Graphs:** Graphs with no loops or multiple edges.

‚û°Ô∏è **Shortest Path:** The minimum number of edges (or minimum weight sum) between two nodes.

‚û°Ô∏è **Graph Distance:** The number of edges in the shortest path.


````python
from collections import deque

def bfs_shortest_path(graph, start, end):
    """Find the shortest path using BFS."""
    visited = set() # Set to keep track of visited nodes
    queue = deque([[start]]) # Create a queue for BFS
    
    while queue: # Loop until the queue is empty
        path = queue.popleft() # Get the first path from the queue
        node = path[-1] # Get the last node from the path
        
        if node == end: # Check if we have found the destination
            return path 
        
        if node not in visited: # Check if the node is already visited
            visited.add(node) # Mark the node as visited
            for neighbor in graph[node]: # Add neighbors of the node to the queue
                new_path = list(path) # Copy the current path
                new_path.append(neighbor) # Add the neighbor to the path
                queue.append(new_path) # Add the new path to the queue
    return None

shortest_path = bfs_shortest_path(G, 0, 33) 
print("Shortest Path from Node 0 to Node 33:", shortest_path)
````

````python
def average_shortest_path_length(graph):
    """Calculate the average shortest path length in a connected graph."""
    total_path_length = 0
    num_paths = 0
    
    for node in graph:
        for target in graph:
            if node != target:
                path = bfs_shortest_path(graph, node, target)
                if path:
                    total_path_length += len(path) - 1
                    num_paths += 1
    
    if num_paths == 0:
        return float('inf')
    
    return total_path_length / num_paths

avg_shortest_path = average_shortest_path_length(adjacency_list)
print("Average Shortest Path Length:", avg_shortest_path)

````

````python
def longest_path_length(graph):
    """
    Find the longest shortest path length in an undirected graph.
    """
    max_length = 0  
    longest_path = []  
    
    nodes = list(graph.keys())
    
    for i in range(len(nodes)):
        for j in range(i + 1, len(nodes)):
            start, end = nodes[i], nodes[j]
            path = bfs_shortest_path(graph, start, end)
            
            if path:
                path_length = len(path) - 1  
                if path_length > max_length:
                    max_length = path_length
                    longest_path = path
    
    return max_length, longest_path

longest_length, longest_path = longest_path_length(adjacency_list)
print("Longest Path Length:", longest_length)
print("Longest Path:", longest_path)
````

````python
def graph_diameter(graph):
    """Calculate the graph diameter using BFS."""
    max_distance = 0 # Initialize the maximum distance to 0
    for node in graph: # Iterate over all nodes in the graph
        for target in graph: 
            if node != target: # Check if the nodes are different
                path = bfs_shortest_path(graph, node, target) # Find the shortest path
                if path: 
                    max_distance = max(max_distance, len(path) - 1) # Update the maximum distance
    return max_distance

diameter = graph_diameter(adjacency_list) 
print("Graph Diameter:", diameter)
````

##### üîó **Connected Components**

A **Connected Component** is a subset of a graph where:
- Every pair of nodes in the subset is **reachable from each other** through a sequence of edges.
- No node in the subset is connected to any node outside the subset.

 üß† **Intuition**

- **Connected Graph:** Has exactly **one connected component** (all nodes are reachable from any node).  
- **Disconnected Graph:** Contains **multiple connected components**, each forming an isolated subgraph. 

In simpler terms:
- If you start from any node in a connected component, you can reach every other node in that component **without leaving it**.



**Largest Connected Component:** In large networks, analyzing the largest connected component often reveals the graph's main structure.  

üõ†Ô∏è **Mathematical Representation**

For a graph \( G = (V, E) \):
- **V:** Set of vertices (nodes).  
- **E:** Set of edges (connections).  

If two nodes \( u \) and \( v \) are in the **same connected component**, there exists a **path** between them:
$$
\text{Path}(u, v) \neq \emptyset
$$

**Number of Connected Components (k):** The total count of these isolated subgraphs.


````python
def dfs(graph, start, visited=None):
    """Depth-First Search for graph traversal."""
    if visited is None: # Initialize the visited nodes set
        visited = set() # Create a set to store visited nodes
    visited.add(start) # Mark the current node as visited
    for neighbor in graph[start]: # Visit all neighbors of the current node
        if neighbor not in visited: # Check if the neighbor is not visited
            dfs(graph, neighbor, visited) # Recursive DFS call
    return visited

def is_connected(graph):
    """Check if the graph is connected."""
    start_node = list(graph.keys())[0] # Start from the first node
    visited = dfs(graph, start_node) # Perform DFS traversal
    return len(visited) == len(graph) # Check if all nodes are visited

connected = is_connected(adjacency_list)
print("Is the Graph Connected?", connected)
````

In case it returned False, it would be interesting to examine also each subgraph:

````python
def find_connected_components(graph):
    """Find all connected components in the graph."""
    visited = set()
    components = []
    
    def dfs(node, component):
        """Recursive DFS to visit all nodes in a component."""
        visited.add(node)
        component.append(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs(neighbor, component)
    
    for node in graph:
        if node not in visited:
            component = []
            dfs(node, component)
            components.append(component)
    
    return components

components = find_connected_components(adjacency_list)
num_components = len(components)
print("Number of Connected Components:", num_components)
print("Connected Components:", components)
````

##### üìà **Clustering Coefficient**

‚û°Ô∏è **Definition:** Measures how well a node's neighbors are connected to each other.

‚û°Ô∏è **Local Clustering Coefficient:** Measures clustering for an individual node.

‚û°Ô∏è **Global Clustering Coefficient:** Measures the overall graph clustering.

‚û°Ô∏è **Importance:** Higher clustering indicates tightly knit groups or communities.

‚û°Ô∏è **Formula:**  


$C_i = \frac{2 * E_i}{k_i * (k_i - 1)}$  
Where:  
- $E_i$ = Number of edges between neighbors of node $i$.  
- $k_i$ = Degree of node $i$.


````python
def clustering_coefficient(graph, node):
    """Calculate local clustering coefficient for a node."""
    neighbors = graph[node] # Get the neighbors of the node
    if len(neighbors) < 2: # Check if the node has less than 2 neighbors
        return 0.0 
    
    actual_edges = sum(1 for i in neighbors for j in neighbors if i != j and j in graph[i]) # checking the neighbors of the node and their connections
    possible_edges = len(neighbors) * (len(neighbors) - 1) # Maximum number of edges between neighbors
    return actual_edges / possible_edges # Calculate the clustering coefficient

# Example Usage
node_clustering = {node: clustering_coefficient(adjacency_list, node) for node in adjacency_list}
print("Clustering Coefficients:", node_clustering)
````

##### üìà **Global Clustering Coefficient**

The **Global Clustering Coefficient** measures the overall tendency of nodes to form **triangles** in a graph. It provides insight into how interconnected the graph is on a global scale.


- **Triangle:** A set of three nodes where **each pair is connected**.  
- **Triplet:** Three nodes connected by **two edges** (forming an open or closed triangle).  
- **Global Clustering Coefficient:** Measures the **fraction of triplets that are closed (form a triangle)**.

üìä **Mathematical Formula**

$$
C = \frac{3 \times \text{Number of Triangles}}{\text{Number of Connected Triplets}}
$$

Where:  
- **Number of Triangles:** Total number of fully connected triples of nodes.  
- **Number of Connected Triplets:** Total sets of three nodes where at least two nodes are connected.

**Why multiply by 3?**  
Each triangle is counted three times (once from each vertex) when counting triplets, so we scale by 3 to correct this.

````python
def count_triangles(graph):
    """Count the total number of triangles in the graph."""
    triangles = 0
    for node in graph: # Iterate over all nodes in the graph
        neighbors = graph[node]
        for i in range(len(neighbors)): # Iterate over all pairs of neighbors
            for j in range(i + 1, len(neighbors)): # Avoid duplicate pairs
                if neighbors[j] in graph[neighbors[i]]: # Check if the pair forms a triangle by having a common neighbor
                    triangles += 1
    return triangles // 3  # Each triangle is counted 3 times

def count_triplets(graph):
    """Count the total number of connected triplets in the graph."""
    triplets = 0
    for node in graph: 
        degree = len(graph[node]) 
        if degree >= 2: 
            triplets += degree * (degree - 1) / 2  # Combination formula nC2
    return triplets

def global_clustering_coefficient(graph):
    """Calculate the global clustering coefficient."""
    triangles = count_triangles(graph)
    triplets = count_triplets(graph)
    if triplets == 0:
        return 0.0
    return (3 * triangles) / triplets 

global_clustering = global_clustering_coefficient(adjacency_list)
print("Global Clustering Coefficient:", global_clustering)
````

##### üèÜ **Centrality Measures**

‚û°Ô∏è **Degree Centrality:** Measures node importance based on the number of connections.  
- **Formula:** $C_D(v) = deg(v)$

‚û°Ô∏è **Closeness Centrality:** Measures how easily a node can reach others in the graph.  
- **Formula:** $C_C(v) = \frac{1}{\sum d(v, u)}$

‚û°Ô∏è **Betweenness Centrality:** Measures how often a node appears on shortest paths between other nodes.  
- **Formula:** $C_B(v) = \sum \frac{\sigma_{st}(v)}{\sigma_{st}}$  
  - $\sigma_{st}$ = Number of shortest paths from $s$ to $t$.  
  - $\sigma_{st}(v)$ = Number of shortest paths through node $v$.

‚û°Ô∏è **Importance:** Centrality helps identify **key influencers, bridges, and critical nodes** in a network.

````python
def degree_centrality(graph):
    """Calculate degree centrality."""
    num_nodes = len(graph)
    return {node: len(neighbors) / (num_nodes - 1) for node, neighbors in graph.items()}

# Example Usage
centrality = degree_centrality(adjacency_list)
print("Degree Centrality:", centrality)

````

````python
def closeness_centrality(graph):
    """Calculate closeness centrality for all nodes in the graph."""
    centrality = {}
    num_nodes = len(graph)
    
    for node in graph:
        total_distance = 0
        reachable_nodes = 0
        
        for target in graph:
            if node != target:
                path = bfs_shortest_path(graph, node, target)
                if path:
                    total_distance += len(path) - 1
                    reachable_nodes += 1
        
        if reachable_nodes > 0 and total_distance > 0:
            centrality[node] = (reachable_nodes) / total_distance
        else:
            centrality[node] = 0.0  # If the node is isolated or unreachable
    
    return centrality


closeness = closeness_centrality(adjacency_list)
print("Closeness Centrality:", closeness)
````

````python
from itertools import permutations

def betweenness_centrality(graph):
    """Calculate betweenness centrality."""
    centrality = {node: 0 for node in graph}
    
    for start, end in permutations(graph.keys(), 2):
        paths = bfs_shortest_path(graph, start, end)
        if paths:
            for node in paths[1:-1]:
                centrality[node] += 1
                
    # Normalize
    num_nodes = len(graph)
    for node in centrality:
        centrality[node] /= (num_nodes - 1) * (num_nodes - 2)
    
    return centrality

# Example Usage
betweenness = betweenness_centrality(adjacency_list)
print("Betweenness Centrality:", betweenness)
````

üîç **Introduction to Graph Search**

Graph search algorithms are used to systematically **explore the vertices and edges** of a graph. They are essential in solving problems related to:
- **Connectivity:** Determining whether two vertices are connected.
- **Shortest Path:** Finding the shortest route between nodes.
- **Cycle Detection:** Identifying loops in a graph.
- **Topological Sorting:** Sequencing tasks or processes.
- **Network Analysis:** Analyzing social or transport networks.

The two most fundamental graph search algorithms are:
- üåê **Breadth-First Search (BFS)**  
- üîó **Depth-First Search (DFS)**  


### üåê **Breadth-First Search (BFS)**

‚úÖ **Definition**
- BFS explores the graph **level by level**, starting from a **source vertex (s)** and visiting all its **neighbors** before moving to their neighbors.  
- It uses a **Queue (FIFO - First In, First Out)** structure.

‚úÖ **Algorithm (High-Level Steps)**  
1. Start from a **source vertex (s)**.  
2. Mark it as **visited** and enqueue it.  
3. While the queue is not empty:  
   - Dequeue a vertex.  
   - Visit all its **unvisited neighbors** and enqueue them.  
4. Repeat until all reachable vertices are visited. 

##### Psuedo Code for Breadth-First Search (BFS)

Input:

    Graph G = (V, E) as an adjacency list
    Source node s

Output:

    Distance from s to each reachable node
    Previous nodes for path reconstruction

BFS:

    Procedure BFS(G, s):

            Initialize:
                dist[u] ‚Üê ‚àû for all nodes u ‚àà V     # Tentative distances
                dist[s] ‚Üê 0                         # Distance to source is 0
                prev[u] ‚Üê None for all u ‚àà V        # To reconstruct paths
                Queue Q = []                        # FIFO queue
            
            Enqueue s into Q                        # Start from the source node

            While Q is not empty:
                u ‚Üê Q.dequeue()                     # Dequeue the front node

                For each neighbor v of u:
                    If dist[v] == ‚àû:                # Unvisited node
                        dist[v] ‚Üê dist[u] + 1       # Update distance
                        prev[v] ‚Üê u                 # Track parent
                        Q.enqueue(v)                # Enqueue neighbor

            Return dist, prev                     

Time Complexity: $O(V+E)$

### üîó ** Depth-First Search (DFS)**

 ‚úÖ **Definition**
- DFS explores the graph by **going as deep as possible** along one branch before **backtracking**.  
- It uses a **Stack (LIFO - Last In, First Out)** structure or recursion.

‚úÖ **Algorithm (High-Level Steps)**  
1. Start from a **source vertex (s)**.  
2. Mark it as **visited**.  
3. Visit an **unvisited neighbor** and recursively explore it.  
4. Backtrack when no unvisited neighbors are left.  
5. Repeat until all vertices are visited.  



##### Pseudo Code for Depth-First Search (DFS) RECURSIVE

Input:

    Graph G = (V, E) as an adjacency list
    Source node s

Output:

    Set of visited nodes
    Previous nodes for path reconstruction

DFS_Recursive: 

    Procedure DFS_Recursive(G, s):

            Initialize:
                visited ‚Üê Empty set                  # Tracks visited nodes
                prev ‚Üê Empty dictionary              # To reconstruct paths
            
            Function Explore(u):
                Mark u as visited
                For each neighbor v of u:
                    If v is not in visited:
                        prev[v] ‚Üê u                  # Track parent
                        Explore(v)                   # Recursive call

            Explore(s)                               # Start DFS from source
            Return visited, prev

##### Pseudo Code for Depth-First Search (DFS) ITERATIVE

Input:

    Graph G = (V, E) as an adjacency list
    Source node s

Output:

    Set of visited nodes
    Previous nodes for path reconstruction
DFS_Iterative:

    Procedure DFS_Iterative(G, s):

            Initialize:
                visited ‚Üê Empty set                 # Tracks visited nodes
                prev ‚Üê Empty dictionary             # To reconstruct paths
                Stack S = []                        # LIFO stack
            
            Push s onto S                           # Start DFS from source

            While S is not empty:
                u ‚Üê S.pop()                         # Process the most recent node
                
                If u is not in visited:
                    Mark u as visited
                    For each neighbor v of u:
                        If v is not in visited:
                            prev[v] ‚Üê u             # Track parent
                            S.push(v)               # Push neighbor onto stack

            Return visited, prev

Time Complexity: $O(V+E)$ for both

üìä **BFS vs DFS: Comparison Table**

| **Feature**        | **BFS**                   | **DFS**                   |
|---------------------|---------------------------|---------------------------|
| **Data Structure** | Queue (FIFO)             | Stack (LIFO) or Recursion |
| **Traversal Style** | Level-by-Level            | Deep-first Traversal      |
| **Shortest Path**  | Yes, in unweighted graphs | No                        |
| **Cycle Detection** | Less efficient            | Efficient                 |
| **Best For**       | Shortest Path, Broadcasting| Topological Sort, Cycles  |
| **Time Complexity** | O(m + n)                 | O(m + n)                 |
| **Space Complexity**| O(n)                     | O(n)                     |


### üöÄ **Dijkstra's Algorithm**

**Dijkstra's Algorithm** is a **graph search algorithm** used to find the **shortest paths** from a single **source node** to all other nodes in a **weighted graph** where **all edge weights are non-negative**.  

It extends the ideas of **Breadth-First Search (BFS)** to handle graphs where edges have **non-uniform weights**.

**Core Idea:** At each step, the algorithm selects the **vertex with the smallest tentative distance** and explores its neighbors, updating their shortest path estimates if a better path is found.

**Key Use Cases:**  
- Routing and navigation systems (e.g., GPS).  
- Network optimization.  
- Flight route planning.  


##### üîÑ **Adaptation of Breadth-First Search (BFS)**

- **BFS Limitation:** BFS assumes **uniform edge weights** (e.g., every edge has a weight of 1).  
- **Dijkstra's Adaptation:** It adapts BFS by introducing a **priority queue** to handle **varying edge lengths**.

‚úÖ **Key Adaptation Concepts:**  
1. **Priority-Based Exploration:** Nodes are processed based on their current shortest path estimate.  
2. **Dummy Nodes (Graph Transformation):** For graphs with **non-uniform edge weights**, they can be transformed into a graph with **unit-length edges** by introducing **dummy nodes** along each edge.

**Intuition:** Instead of traversing edges uniformly, Dijkstra uses a **priority mechanism** to prioritize vertices with the **smallest accumulated distance**.


##### üõ†Ô∏è **Priority Queue Implementation**

Dijkstra's algorithm relies heavily on an **efficient priority queue** to select the next node to process.

- Quickly retrieves the node with the **smallest tentative distance**.  
- Efficiently updates the shortest path estimates for neighbors.

‚úÖ **Common Priority Queue Implementations:**
| **Data Structure** | **Insertion** | **Extract Min** | **Update Key** |
|---------------------|---------------|-----------------|---------------|
| **Array**          | O(1)          | O(n)            | O(n)          |
| **Binary Heap**    | O(log n)      | O(log n)        | O(log n)      |
| **d-ary Heap**     | O(log_d n)    | O(log_d n)      | O(log_d n)    |
| **Fibonacci Heap** | O(1)          | O(log n)        | O(1)          |

‚úÖ **Which is Best?**
**Binary Heap:** Standard choice for balanced efficiency

**Intuition:** The priority queue helps ensure that nodes are visited in the correct order of increasing shortest path estimates.

The efficiency of Dijkstra's Algorithm depends heavily on the **choice of the priority queue**.


##### Pseudo Code for Dijkstra's Algorithm
Input:

    Graph G = (V, E) as an adjacency list
    Edge weights w(u, v) > 0
    Source node s

Output:

    Shortest distances from s to all nodes
    Previous nodes to reconstruct the paths

Dijkstra:

    Procedure Dijkstra(G, s):

            Initialize:
                dist[u] ‚Üê ‚àû for all nodes u ‚àà V     # Tentative distances
                dist[s] ‚Üê 0                         # Distance to source is 0
                prev[u] ‚Üê None for all u ‚àà V        # To reconstruct paths
                PriorityQueue Q = []                # Min-heap: (distance, node)

            Insert (0, s) into Q                    # Start from the source node

            While Q is not empty:
                (current_dist, u) ‚Üê Q.pop()         # Node with smallest distance

                For each neighbor v of u:
                    weight ‚Üê w(u, v)                # Edge weight
                    distance ‚Üê current_dist + weight
                    
                    If distance < dist[v]:          # Found a shorter path
                        dist[v] ‚Üê distance
                        prev[v] ‚Üê u
                        Q.push((distance, v))       # Update priority queue

            Return dist, prev                     


Time Complexity:  $O(V+E)logV)$ using a binary heap.


### Our Implementation of Girvan-Newman
The Girvan-Newman algorithm is a community detection method for graphs that works by progressively removing edges with the highest betweenness centrality. It identifies communities (connected subgraphs) by breaking the graph apart iteratively.

1. Calculate Edge Betweenness Centrality: Betweenness measures how often an edge appears on the shortest paths between nodes.
We can use either:

- BFS: Efficient for unweighted graphs.

- Dijkstra: General-purpose for weighted graphs.

2. Remove the Edge with Highest Betweenness: The edge with the largest centrality score is removed as it is considered a "bridge" between communities.

4. Find Connected Components: After edge removal, the graph is split into components (disconnected subgraphs).
This can be done using either using
- DFS (Iterative)
- DFS (Recursive) methods.

Repeat: recalculating betweenness and removing edges until the desired number of communities is reached or no edges remain.

````python
# Girvan-Newman Algorithm for Community Detection

import heapq
from collections import defaultdict, deque
import matplotlib.pyplot as plt
import numpy as np
import copy
import networkx as nx   
import pandas as pd
import itertools

def find_connected_components(graph, method="dfs_recursive"):
    """
    Find all connected components in an undirected graph using DFS.

    Args:
        graph (dict): Adjacency list representing the graph.
        method (str): 'dfs_recursive' or 'dfs_iterative'.

    Returns:
        list: List of connected components.
    """
    visited = set()
    components = []

    # Recursive version
    # Source: https://www.geeksforgeeks.org/depth-first-search-or-dfs-for-a-graph/
    def dfs_recursive(node, component):
        """
        Recursive Depth-First Search (DFS) function to explore a connected component in the graph.
        
        This function visits all nodes reachable from the given 'node' and adds them to the same 
        connected component. It uses recursion to traverse all unvisited neighbors.
        """
        visited.add(node)  # Mark the current node as visited
        component.append(node)  # Add the current node to the component
        
        #print(f"Step {step_counter}:")
        #print(f"  Current Node: {node}")
        #print(f"  Visited Nodes: {visited}")
        #print(f"  Current Component: {component}")
        #print("----------------------------------")
        #step_counter += 1
    
        # Recursively visit all unvisited neighbors
        for neighbor in graph[node]:
            if neighbor not in visited:  # Process only unvisited nodes
                dfs_recursive(neighbor, component)

    # Iterative version
    def dfs_iterative(start):
        """
        Iterative Depth-First Search using an explicit stack.
        """
        stack = [start]
        component = []  # Local list to store the component nodes

        while stack:
            node = stack.pop()  # Pop the most recently added node (LIFO)
            if node not in visited:
                visited.add(node)
                component.append(node)
                
                # Add unvisited neighbors to the stack
                for neighbor in reversed(graph[node]):
                    if neighbor not in visited:
                        stack.append(neighbor)
        return component

    # Which one to use? Both give same results, O notation is the same but Recursive is more intuitive but may have some problem in larger dataset due to recursion limit that on Py is around 1000 calls

    # Explore all nodes
    for node in graph:
        if node not in visited:
            if method == "dfs_recursive":
                component = []
                dfs_recursive(node, component)
                components.append(component)
            elif method == "dfs_iterative":
                component = dfs_iterative(node)
                components.append(component)
            else:
                raise ValueError("Invalid method. Use 'dfs_recursive' or 'dfs_iterative'.")

    return components


# Source: https://www.geeksforgeeks.org/breadth-first-search-or-bfs-for-a-graph/
def bfs_shortest_paths(graph, source):
    """
    Perform Breadth-First Search (BFS) to calculate the shortest paths from a given source node to all other nodes.
    BFS explores vertices level by level, using a queue data structure. This implementation also tracks the parents 
    of each node to reconstruct shortest paths.
    """
    queue = deque([source]) # Initialize a queue for BFS, starting with the source node

    distances = {node: float('inf') for node in graph} # Set the distance to all nodes as infinity (unvisited)
    distances[source] = 0 # Set the distance to the source node as 0 (starting point).

    # Initialize the parents dictionary:
    # - Tracks all parents for each node along shortest paths.
    parents = defaultdict(list) # parents dict tracks all parents for each node along shortest paths
    
    #step = 0

    while queue: # until all vertices are reachable
        # Dequeue the current node
        current = queue.popleft() # remove from queue

        #print(f"Step {step}:")
        #print(f"  Current Node: {current}")
        #print(f"  Queue: {list(queue)}")
        #print(f"  Distances: {distances}")
        #print(f"  Parents: {dict(parents)}")

        # Explore all adjacent vertices of the current node
        for neighbor in graph[current]:
            if distances[neighbor] == float('inf'): # so not visited
                distances[neighbor] = distances[current] + 1 # update as one level deeper than current node
                queue.append(neighbor) # add to queue to process later

            if distances[neighbor] == distances[current] + 1: # if distance is exactly one level deeper than the current node
                parents[neighbor].append(current) # Add the current node as a parent of the neighbor

    return distances, parents

# Source: https://symbio6.nl/en/blog/analysis/betweenness-centrality (only for the concept)
from collections import defaultdict

def calculate_betweenness_bfs(graph):
    """
    Calculate edge betweenness centrality for all edges in an undirected network.

    Betweenness centrality measures the "bridgeness" of an edge by evaluating the fraction
    of shortest paths between all pairs of nodes in the network that pass through that edge.
    """

    betweenness = defaultdict(int) # Dict to store BC values for edges

    for node in graph:
        #print(f"\nProcessing Source Node: {node}")
        distances, parents = bfs_shortest_paths(graph, node) # BFS return shortest paths and their parent relationships
        #print(f"  BFS Distances from {node}: {distances}")
        #print(f"  BFS Parents from {node}: {dict(parents)}")
        node_flow = {n: 1 for n in graph} # Initialize flow for all nodes as 1 (default contribution to shortest paths)

        nodes_by_distance = sorted(distances, key=distances.get, reverse=True) # Process nodes with farthest nodes first
        #print(f"  Nodes by Distance (Reverse): {nodes_by_distance}")

        # Backtrack from farthest nodes to distribute flow across edges
        for target in nodes_by_distance: 
            for parent in parents[target]:
                # Define the edge between parent and target
                edge = tuple(sorted((parent, target)))  # Define edge between parent and target. Sort to avoid duplicates
                
                # Distribute flow proportionally across all shortest paths to the target
                flow = node_flow[target] / len(parents[target])  # Split flow equally
                betweenness[edge] += flow  # Add flow to the edge's betweenness
                node_flow[parent] += flow  # Pass flow back to the parent node
                #print(f"    Updated Betweenness for Edge {edge}: {betweenness[edge]}")
                #print(f"    Flow Distribution: Node {parent} Flow: {node_flow[parent]}")
                
    #for edge, score in betweenness.items():
        #print(f"  Edge {edge}: {score}")

    return betweenness

# Dijkstra's Algorithm 
def dijkstra_adj_list(graph, start_node):
    """
    Dijkstra's Algorithm to find the shortest paths from a start node
    in an adjacency list representation of a graph.
    """
    distances = {node: float('inf') for node in graph}
    distances[start_node] = 0
    previous_nodes = {node: None for node in graph}
    priority_queue = [(0, start_node)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)
        for neighbor in graph[current_node]:
            weight = 1
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous_nodes[neighbor] = current_node
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances, previous_nodes

# Betweenness Centrality Calculation: Dijkstra based
def calculate_betweenness_dijkstra(graph):
    """
    Calculate edge betweenness centrality using Dijkstra's Algorithm.
    """
    betweenness = defaultdict(float)

    for node in graph:
        #print(f"\n[DEBUG] Running Dijkstra from source node: {node}")
        distances, parents = dijkstra_adj_list(graph, node)
        node_flow = {n: 1 for n in graph}

        # Sort nodes by distance from the source (farthest nodes processed first)
        nodes_by_distance = sorted(distances, key=distances.get, reverse=True)
        #print(f"[DEBUG] Nodes by distance (farthest to closest): {nodes_by_distance}")

        for target in nodes_by_distance:
            if parents[target] is not None:
                edge = tuple(sorted((parents[target], target)))  # Avoid duplicates
                flow = node_flow[target] / 1  # In unweighted graphs, flow = 1
                betweenness[edge] += flow
                node_flow[parents[target]] += flow
                #print(f"[DEBUG] Edge {edge} receives flow: {flow:.2f}")

 #   print("\n[DEBUG] Final Betweenness Centrality:")
  #  for edge, centrality in betweenness.items():
   #     print(f"  Edge {edge}: {centrality:.2f}")

    return betweenness
    
# Source: https://memgraph.github.io/networkx-guide/algorithms/community-detection/girvan-newman/
# Girvan-Newman Algorithm
def custom_girvan_newman(graph, betweenness_method="bfs", component_method="dfs_recursive"):
    """
    Girvan-Newman algorithm for community detection.
    Args:
        graph (dict): Input graph as adjacency list.
        betweenness_method (str): 'bfs' or 'dijkstra' for edge betweenness.
        component_method (str): 'dfs_recursive' or 'dfs_iterative' for connected components.
    Returns:
        tuple: (communities, removed_edges)
    """
    graph_copy = copy.deepcopy({node: list(neighbors) for node, neighbors in graph.adjacency()})
    removed_edges = []
    iteration = 0

    while True:
        print(f"\n[DEBUG] Iteration {iteration}: Calculating betweenness centrality...")

        # Calculate betweenness centrality with choosed method
        if betweenness_method == "bfs":
            betweenness = calculate_betweenness_bfs(graph_copy)
        elif betweenness_method == "dijkstra":
            betweenness = calculate_betweenness_dijkstra(graph_copy)
        else:
            raise ValueError("Invalid betweenness method. Choose 'bfs' or 'dijkstra'.")

        # Stop if no edges remain
        if not betweenness:
            break

        # Remove the edge with the highest betweenness
        edge_to_remove = max(betweenness, key=betweenness.get)
        graph_copy[edge_to_remove[0]].remove(edge_to_remove[1])
        graph_copy[edge_to_remove[1]].remove(edge_to_remove[0])
        removed_edges.append(edge_to_remove)
        print(f"[DEBUG] Removed edge: {edge_to_remove}")

        # Find connected components
        communities = find_connected_components(graph_copy, method=component_method)
        print(f"[DEBUG] Communities: {communities}")

        # Terminate when more than one community is detected
        if len(communities) > 1:
            break

        iteration += 1

    return communities, removed_edges

# Visualization
def visualize_communities(graph, communities, removed_edges=[]):
    """
    Visualize the graph and its communities.
    """
    # Assign colors to nodes based on their community
    colors = plt.cm.rainbow(np.linspace(0, 1, len(communities)))
    color_map = {}
    for i, community in enumerate(communities):
        for node in community:
            color_map[node] = colors[i]

    pos = nx.spring_layout(graph)

    plt.figure(figsize=(10, 10))
    nx.draw_networkx_nodes(graph, pos, node_size=500, 
                           node_color=[color_map[node] for node in graph.nodes()],
                           edgecolors='black')
    nx.draw_networkx_labels(graph, pos, font_size=12, font_color='white')

    # Draw edges excluding removed edges
    edges_to_draw = [(u, v) for u, v in graph.edges() if (u, v) not in removed_edges and (v, u) not in removed_edges]
    nx.draw_networkx_edges(graph, pos, edgelist=edges_to_draw, width=2, alpha=0.5)

    plt.title("Girvan-Newman Communities")
    plt.axis("off")
    plt.show()
````

# Optimization Techniques for Community Detection

##### Spectral Clustering for Community Detection

**Input**:

- Graph \( G = (V, E) \) represented as an adjacency matrix.
- Number of clusters \( k \).

**Output**:

- Detected communities.

**Procedure**:

**Compute adjacency and degree matrices**:

    - \( \text{adj\_matrix} \) is the adjacency matrix of \( G \).
    - \( \text{degrees}[i][i] = \sum_j \text{adj\_matrix}[i][j] \) (diagonal degree matrix).

**Calculate the normalized Laplacian**:

    - Compute \( \text{D\_inv\_sqrt}[i][i] = \frac{1}{\sqrt{\text{degrees}[i][i]}} \).
    - Replace \( NaN \) values with \( 0 \) (to handle division by zero).
    - Compute the normalized Laplacian:
      \( L = I - \text{D\_inv\_sqrt} \times \text{adj\_matrix} \times \text{D\_inv\_sqrt} \).

**Perform eigenvalue decomposition**:

    - Compute eigenvalues and eigenvectors of \( L \).
    - Sort eigenvalues in ascending order.
    - Select eigenvectors corresponding to the \( k \) smallest non-trivial eigenvalues.

**Normalize the eigenvectors**:

    - Normalize each row of the eigenvector matrix.

**Apply K-means clustering**:

    - Use the \( k \)-dimensional eigenvector representation of nodes as input to K-means clustering.
    - Assign each node to a cluster based on the K-means result.

**Group nodes into communities**:

    - For each cluster, group the nodes that belong to it.

**Return**:

    - Detected communities.


````python
# Graph Class (Adjacency List Implementation)
class Graph:
    def __init__(self, edges):
        self.graph = defaultdict(set)
        self.edges = edges
        self.nodes = set()
        for u, v in edges:
            self.add_edge(u, v)

    def add_edge(self, u, v):
        self.graph[u].add(v)
        self.graph[v].add(u)
        self.nodes.update([u, v])

    def degree(self, node):
        return len(self.graph[node])

    def adjacency_matrix(self):
        nodes = sorted(self.nodes)
        n = len(nodes)
        node_idx = {node: i for i, node in enumerate(nodes)}
        adj_matrix = np.zeros((n, n))
        for u in nodes:
            for v in self.graph[u]:
                adj_matrix[node_idx[u], node_idx[v]] = 1
        return adj_matrix, nodes
    
    def adjacency(self):
        for node, neighbors in self.graph.items():
            yield node, list(neighbors)

def calculate_modularity(adj_matrix, communities):
    """
    It evaluates the modularity of the resulting communities.
    """
    m = np.sum(adj_matrix) / 2
    Q = 0.0
    for community in communities:
        # For every pair of nodes in the same community i,j
        for i in community:
            for j in community:
                A_ij = adj_matrix[i, j] # Adjacency matrix value
                k_i = np.sum(adj_matrix[i]) # Degree of node i
                k_j = np.sum(adj_matrix[j]) # Degree of node j
                Q += A_ij - (k_i * k_j) / (2 * m) # Modularity formula numerator
    return Q / (2 * m) # Normalize by 2m
````

````python
from sklearn.preprocessing import normalize
import numpy as np
from collections import defaultdict

# Source: https://rahuljain788.medium.com/implementing-spectral-clustering-from-scratch-a-step-by-step-guide-9643e4836a76
def spectral_clustering(graph, k):
    """
    Spectral Clustering on a graph using normalized Laplacian and custom K-means.
    """
    # Adjacency and Degree Matrices
    adj_matrix, nodes = graph.adjacency_matrix()
    degrees = np.diag(adj_matrix.sum(axis=1))

    # Normalized Laplacian
    D_inv_sqrt = np.diag(1.0 / np.sqrt(np.diag(degrees)))
    D_inv_sqrt = np.nan_to_num(D_inv_sqrt)  # Handle division by zero
    L = np.eye(len(nodes)) - D_inv_sqrt @ adj_matrix @ D_inv_sqrt

    # Eigenvalue Decomposition
    eigenvalues, eigenvectors = np.linalg.eigh(L)
    sorted_indices = np.argsort(eigenvalues)  # Sort eigenvalues
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]
    
    print("Sorted Eigenvalues:", eigenvalues[:k+1])

    # Select and normalize eigenvectors corresponding to smallest non-trivial eigenvalues
    k_smallest_eigenvectors = eigenvectors[:, 1:k+1]  # Skip the first trivial eigenvector
    k_smallest_eigenvectors = normalize(k_smallest_eigenvectors, axis=1)

    # Run K-Means Clustering
    centroids, cluster_assignments = kmeans(k_smallest_eigenvectors, k, method="kmeans++", seed=42)

    # Group Nodes into Communities
    communities = defaultdict(list)
    for i, cluster_id in enumerate(cluster_assignments):
        communities[cluster_id].append(nodes[i])

    return list(communities.values())
````

##### Louvain Method for Community Detection

**Input**:

- Graph \( G = (V, E) \) represented as an adjacency matrix.
- Maximum number of iterations \( \text{max\_iter} \).

**Output**:

- Detected communities.
- Final modularity score.

**Procedure**:

**Initialize**:

    - Let \( n \) be the number of nodes in \( G \).
    - Compute \( \text{degrees}[i] = \sum_{j} G[i][j] \) for all nodes \( i \).
    - Set \( \text{inv\_m} = \frac{1}{\sum \text{degrees}} \) for modularity normalization.
    - Assign each node to its own community:
      \( \text{communities}[i] = i \).

**Define modularity gain**:

    - \( k_i = \text{degrees}[i] \).
    - Let \( \text{delta\_q} = 0 \).
    - For each neighbor \( j \) of node \( i \):
        - If \( j \) belongs to \( \text{target\_comm} \), increment \( \text{delta\_q} \) by \( G[i][j] \).
        - If \( j \) belongs to \( \text{curr\_comm} \), decrement \( \text{delta\_q} \) by \( G[i][j] \).
    - Compute the degree sums for the target and current communities:
        - \( \text{sum\_target} = \sum \text{degrees}[k] \, \forall k \in \text{target\_comm} \).
        - \( \text{sum\_curr} = \sum \text{degrees}[k] \, \forall k \in \text{curr\_comm} \).
    - Adjust \( \text{delta\_q} \):
        - \( \text{delta\_q} -= k_i \times (\text{sum\_target} - k_i) \times \text{inv\_m} \).
        - \( \text{delta\_q} += k_i \times (\text{sum\_curr} - k_i) \times \text{inv\_m} \).

**Iteratively refine communities**:

    - For \( \text{max\_iter} \) iterations:
        - Set \( \text{moved} = \text{False} \).
        - For each node \( i \):
            - Evaluate the modularity gain for moving \( i \) to each neighboring community.
            - Move \( i \) to the community with the highest modularity gain if it improves the modularity.
            - Update \( \text{moved} \) if any node was reassigned.
        - Terminate if no nodes were moved.

**Extract communities**:

    - Group nodes by their final community assignments.

**Return**:

    - Detected communities.
    - Final modularity score.


````python
# Louvain Algorithm

# Source: https://users.ece.cmu.edu/~lowt/papers/Louvain_accepted.pdf

def louvain_cluster(adj_matrix, max_iter=10):
    """
    Simplified version of the Louvain clustering algorithm using NumPy. 
    The algorithm aims to detect communities in a graph by iteratively optimizing the modularity of the graph. 
    """
    n = adj_matrix.shape[0]  # Number of nodes
    degrees = np.sum(adj_matrix, axis=1)
    inv_m = 1.0 / np.sum(degrees)  # 2m for modularity normalization
    communities = np.arange(n)  # Initialize each node in its own community

    def modularity_gain(node, target_comm, curr_comm):
        """Compute the modularity gain of moving 'node' to 'target_comm'."""
        k_i = degrees[node]
        delta_q = 0.0
        # Direct Contributions
        for neighbor, weight in enumerate(adj_matrix[node]): # It iterates over all neighbors of the node
            if weight > 0:
                # Depending by where the neighbor belongs, add edge weight to the modularity gain
                if communities[neighbor] == target_comm:
                    delta_q += weight
                if communities[neighbor] == curr_comm:
                    delta_q -= weight
        # Indirect Contributions
        # Compute the sum of degrees of nodes in the target and current communities
        sum_in_target = np.sum(degrees[communities == target_comm])
        sum_in_curr = np.sum(degrees[communities == curr_comm])
        # Adjusted formula
        delta_q -= k_i * (sum_in_target - k_i) * inv_m
        delta_q += k_i * (sum_in_curr - k_i) * inv_m
        return delta_q

    # Iterative Community Refinement
    for iteration in range(max_iter): # Either max_iter or until no nodes are moved
        moved = False
        for node in range(n): # for each node
            curr_comm = communities[node]
            max_gain = 0
            best_comm = curr_comm
 
            # Evaluate modularity gain for moving it to each neighboring community
            for neighbor, weight in enumerate(adj_matrix[node]):
                if weight > 0 and communities[neighbor] != curr_comm:
                    target_comm = communities[neighbor]
                    gain = modularity_gain(node, target_comm, curr_comm)
                    if gain > max_gain:
                        max_gain = gain
                        best_comm = target_comm

            # Reassign the node to the best community
            if best_comm != curr_comm:
                communities[node] = best_comm
                moved = True

        if not moved:  # Stop if no nodes were moved
            break

    return extract_communities(communities)

def extract_communities(communities):
    """Group nodes by their community assignments."""
    community_groups = defaultdict(list)
    for node, comm in enumerate(communities):
        community_groups[comm].append(node)
    return list(community_groups.values())
````

#### **1. Lambiotte Coefficient (Node Importance)**
The **Lambiotte Coefficient** measures how central a node is to its community. It evaluates the proportion of connections a node has within its assigned community relative to all possible connections.

- **Formula**:
  
  $L(n) = \frac{\text{Internal Degree of Node } n}{\text{Degree of Node } n}$

  Where:
  - **Internal Degree**: Number of edges the node shares with nodes in its own community.
  - **Degree**: Total number of edges connected to the node.

- **Interpretation**:
  - Values close to **1.0**: The node is strongly integrated into its community and most of its connections are internal.
  - Lower values (e.g., 0.5): The node has significant connections outside its community, making it less central.

---

#### **2. Clauset's Parameter (Community Strength)**
The **Clauset's Parameter** measures the cohesion or density of a community relative to the entire graph. It evaluates how well a community is connected internally versus its connections to other communities.

- **Formula**:
  
  $C(C_i) = \frac{\text{Internal Edges in } C_i}{\text{Total Edges Involving Nodes in } C_i}$
  
  
  Where:
  - **Internal Edges**: Edges within the community.
  - **Total Edges**: All edges involving nodes in the community, including external edges.

- **Interpretation**:
  - Values close to **1.0**: The community is highly cohesive, with most of its edges being internal.
  - Lower values: The community has weaker internal cohesion, with many external connections.


````python
# Additional Metrics

def lambiotte_coefficient(adj_matrix, communities):
    """
    Compute the Lambiotte coefficient for each node.
    """
    n = adj_matrix.shape[0]
    node_importance = {}

    for community in communities: # For each community
        for node in community: # For each node in the community
            k_in = np.sum([adj_matrix[node, neighbor] for neighbor in community]) # Internal degree
            k_total = np.sum(adj_matrix[node])  # Total degree
            L = k_in / k_total if k_total > 0 else 0 # Lambiotte formula # Avoid division by zero
            node_importance[node] = L
    
    return node_importance

def clauset_parameter(adj_matrix, communities):
    """
    Compute Clauset's parameter for each community.
    """
    community_quality = {}

    for idx, community in enumerate(communities): # For each community
        E_in = sum(adj_matrix[i, j] for i in community for j in community if i != j) / 2  # Internal edges
        E_out = sum(adj_matrix[i, j] for i in community for j in range(adj_matrix.shape[0]) if j not in community) # External edges
        
        Q = E_in / (E_in + E_out) if (E_in + E_out) > 0 else 0  # Clauset's formula # Avoid division by zero
        community_quality[idx] = Q

    return community_quality
````

### üõ†Ô∏è **Minimum Spanning Tree (MST)**

- üìù The **Minimum Spanning Tree (MST)** problem focuses on connecting **all vertices** in an undirected graph using the **minimum possible total edge weight**.  
-  The goal is to minimize the total cost of the connections while ensuring all vertices remain connected.  

‚úÖ **Real-World Examples:**
- **Network Design:** Laying cables for telephone or computer networks with minimum cost.  
- **Transportation Planning:** Designing road networks with minimum total distance.  
- **Document Clustering:** Determining similarity networks between documents in machine learning.

√π‚úÖ **Key Requirements of MST:**
1. The graph must be **connected**.  
2. All vertices must be included in the spanning tree.  
3. The total weight of the selected edges must be **minimized**.

---

##### üìê **Graphs Recap**

- üìù A **graph** consists of a set of **vertices (nodes)** and **edges (connections between nodes)**.  
- üìù Graphs can be represented using:
   - **Adjacency List:** Space-efficient for sparse graphs.  
   - **Adjacency Matrix:** Fast edge lookups, better for dense graphs.  
- üìù The focus in MST problems is on **undirected graphs** with **weighted edges**.

 **Graph Representation Example:**

- **Vertices (V):** {A, B, C, D}  
- **Edges (E):** {(A-B, 2), (B-C, 1), (C-D, 3), (A-D, 4)}  

In an MST, the edges chosen would minimize the total weight while keeping all vertices connected

### üå≥ **Spanning Trees**

- üìù A **Spanning Tree** is a **subgraph** of a connected graph that:
   1. Includes **all the vertices** of the graph
   2. Forms a **tree** (no cycles).  
   3. Is **connected**.

- üìù The **Minimum Spanning Tree (MST)** is a spanning tree with the **minimum total edge weight**.

The number of spanning trees in a graph can be estimated using:
$$
\# \text{Spanning Trees} = \binom{|E|}{|V| - 1} - \# \text{Cycles}
$$

Where:
- $ |E| $: Total number of edges in the graph.
- $ |V| $: Total number of vertices in the graph.
- $\# \text{Cycles} $: Total number of independent cycles in the graph.

‚úÖ **Key Properties:**
1. A connected graph will always have **at least one spanning tree**.  
2. A graph with **n vertices** will have at least **n-1 edges** in its spanning tree.  
3. Removing any edge from the spanning tree will **disconnect the graph**.

### üõ†Ô∏è **Prim‚Äôs Algorithm**

- üìù **Prim's Algorithm** is a **greedy algorithm** used to construct the MST by **growing a tree one vertex at a time**.  
- üìù It starts from an **arbitrary vertex** and keeps adding the **minimum-weight edge** that connects a visited vertex to an unvisited vertex.

‚úÖ **Steps of Prim's Algorithm:**
1. Start from any vertex.  
2. Add the **smallest edge** that connects the current tree to an unvisited vertex.  
3. Repeat until all vertices are included in the tree.  

‚úÖ **Key Characteristics:**
- Builds the tree **vertex by vertex**.  
- Uses a **priority queue** (or min-heap) to efficiently retrieve the **minimum-weight edge**.

Best for: Dense graphs (many edges).


````python
import heapq

# Manual Implementation of Prim's Algorithm
def prim_mst(graph):
    """
    Implements Prim's algorithm to find the Minimum Spanning Tree (MST) of a graph.
    Uses a priority queue to always select the smallest edge connecting visited and unvisited nodes.
    """
    mst = nx.Graph()  # Initialize an empty graph for the MST
    visited = set()  # Track visited nodes
    start_node = list(graph.nodes)[0]  # Start from the first node
    visited.add(start_node) # Mark the start node as visited
    
    # Add all edges from the start node to the priority queue
    edges = []
    for neighbor in graph.neighbors(start_node): # Get all neighbors of the start node
        weight = graph[start_node][neighbor].get('weight', 1)  # Default weight is 1 if not present
        edges.append((weight, start_node, neighbor)) # Add edge to the priority queue
    heapq.heapify(edges)  # Convert edges into a priority queue
    
    while edges:
        weight, u, v = heapq.heappop(edges)  # Get the edge with the smallest weight
        if v not in visited: # Process only unvisited nodes
            visited.add(v)  # Mark the new node as visited
            mst.add_edge(u, v, weight=weight)  # Add edge to MST
            
            # Add edges from the newly visited node to the queue
            for neighbor in graph.neighbors(v):
                if neighbor not in visited:
                    weight = graph[v][neighbor].get('weight', 1) 
                    heapq.heappush(edges, (weight, v, neighbor)) # Add edge to the priority queue
    
    return mst
````

### üîó **Kruskal‚Äôs Algorithm**

- üìù **Kruskal's Algorithm** is another **greedy algorithm** used to construct the MST by **selecting edges in increasing order of weight**.  
- üìù It ensures that no **cycles** are formed when adding edges.

‚úÖ **Steps of Kruskal's Algorithm:**
1. Sort all edges in **ascending order** of weight.  
2. Start adding edges to the tree, one at a time, starting from the **smallest weight edge**.  
3. Skip any edge that forms a **cycle**.  
4. Stop when **n-1 edges** have been added.

**Key Characteristics:**
- Builds the tree **edge by edge**.  
- Uses a **Union-Find Data Structure** to check for **cycle formation**.

Best for: Sparse graphs (fewer edges).

````python
# Manual Implementation of Kruskal's Algorithm
def kruskal_mst(graph):
    """
    Implements Kruskal's algorithm to find the Minimum Spanning Tree (MST) of a graph.
    Sorts edges by weight and adds them one by one while avoiding cycles.
    """
    mst = nx.Graph()  # Initialize an empty graph for the MST
    edges = sorted(graph.edges(data=True), key=lambda x: x[2]['weight'])  # Sort edges by weight
    parent = {node: node for node in graph.nodes}  # Initialize Union-Find structure by setting each node as its parent
    
    def find(node):
        """Finds the root of a node using path compression."""
        if parent[node] != node: # If the node is not its parent
            parent[node] = find(parent[node]) # Update the parent by recursion
        return parent[node]
    
    def union(u, v):
        """Joins two subsets into one."""
        root_u = find(u) # Find the root of the first subset
        root_v = find(v) # Find the root of the second subset
        if root_u != root_v: # If the roots are different
            parent[root_u] = root_v # Union the subsets by setting one as the parent of the other
    
    for u, v, attr in edges: # Process edges in ascending order
        if find(u) != find(v):  # Check if adding the edge forms a cycle
            weight = attr.get('weight', 1) 
            mst.add_edge(u, v, weight=weight)  # Add edge to MST
            union(u, v)  # Union the subsets
    
    return mst
````

### ‚öôÔ∏è **2. Randomized Min-Cut Algorithm**

The **Randomized Min-Cut Algorithm** is a probabilistic approach to finding a **minimum cut-set** in a graph.

- A **Cut-Set** is a set of edges in a graph whose removal **disconnects** the graph into two or more components.

üõ†Ô∏è **2.1 Algorithm Explanation**

1. **Graph Contraction:** 
   - Iteratively contract edges selected **uniformly at random**.
   - Replace two connected nodes with a **single "super-node"**, merging their edges.

2. **Iteration Mechanism:** 
   - Start with a graph of \( n \) vertices.
   - Contract edges **n-2 times** until only **two super-nodes** remain.

3. **Key Objective:** 
   - Avoid contracting edges that are part of the **minimum cut-set**.

---

 üîÑ **2.2 Iterative Process**

1. **Start:** Begin with the original graph.
2. **Random Edge Selection:** Uniformly select an edge at random.
3. **Edge Contraction:** Combine the two vertices connected by the selected edge.
4. **Repeat:** Continue contracting edges until only **two vertices** remain.
5. **Result:** The remaining edges represent a **candidate minimum cut-set**.

**Visualization:**  
- Nodes merge during each contraction.  
- The graph simplifies while preserving the critical edges of the minimum cut.

---

üìä **2.3 Probabilistic Analysis**

The algorithm is not **guaranteed** to find the minimum cut in a **single run**. However:

- **Key Probability Insight:** The probability of finding the correct minimum cut in one run is at least:  
$$
P(\text{Correct Min-Cut}) \geq \frac{2}{n(n-1)}
$$

- **Bounded Probability of Error:** Running the algorithm multiple times **increases the confidence** in finding the correct minimum cut.

**Intuition:**  
- With each contraction, the chance of contracting an **edge in the min-cut** reduces.  
- Multiple independent runs reduce the probability of failure exponentially.

````python
import random
import copy

def manual_min_cut(graph):
    """
    Implements the Minimum Cut algorithm manually for a directed weighted graph.
    """
    # Create a local copy of the graph
    local_graph = copy.deepcopy(graph)
    adjacency_list = {node: list(local_graph.neighbors(node)) for node in local_graph.nodes()}
    edge_weights = {(u, v): data.get('capacity', 1) for u, v, data in local_graph.edges(data=True)} 
    
    # Helper function to merge two nodes
    def contract_edge(u, v):
        """Contract an edge between u and v, merging v into u."""
        for neighbor in list(local_graph.neighbors(v)): # Get all neighbors of v
            if neighbor != u:  # Avoid self-loops 
                weight = local_graph[v][neighbor].get('capacity', 1) # Get edge weight
                if local_graph.has_edge(u, neighbor): # If edge already exists:
                    local_graph[u][neighbor]['capacity'] += weight # update the weight
                else:
                    local_graph.add_edge(u, neighbor, capacity=weight) # Add a new edge
        
        # Remove node v and its edges
        local_graph.remove_node(v)
    
    # Run contraction until two nodes remain
    while len(local_graph.nodes()) > 2:
        # Choose a random edge
        u, v = random.choice(list(local_graph.edges())) 
        contract_edge(u, v) # Merge nodes u and v
    
    # Calculate the minimum cut
    nodes = list(local_graph.nodes()) # Get the remaining nodes
    cut_edges = [
        (u, v) for u, v in graph.edges() # Get all original edges
        if (u in nodes[0] and v in nodes[1]) or (u in nodes[1] and v in nodes[0]) # Check if they cross the cut
    ] # Find the edges connecting the two final nodes
    cut_value = sum(edge_weights.get((u, v), 1) for u, v in cut_edges) # Calculate the total cut value by summing edge weights
    
    return cut_value, cut_edges
````
